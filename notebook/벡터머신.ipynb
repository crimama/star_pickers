```python
def minmax(df):
  temp_df = df.reshape(-1,6)
  min = np.min(temp_df,axis=0)
  max = np.max(temp_df,axis=0)
  return min,max
def scailng(df):
  global min,max
  scailed_df = (df-min)/(max-min)
  return scailed_df
```
